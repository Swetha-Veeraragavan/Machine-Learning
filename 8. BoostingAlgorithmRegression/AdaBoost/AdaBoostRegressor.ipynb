```python
import pandas as pd
```

```python
dataset=pd.read_csv("50_Startups.csv")
dataset.head()
```

```python
dataset=pd.get_dummies(dataset,drop_first=True)
dataset.head()
```

```python
dataset.columns
```

```python
independent=dataset[['R&D Spend', 'Administration', 'Marketing Spend','State_Florida', 'State_New York']]
```

```python
dependent=dataset[["Profit"]]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(independent, dependent, test_size=0.30,random_state=0)
```

```python
X_train
```

### Standardize the inputs

```python
'''
Since above method provides less r2_score value, so we need try to standardized and thn cal r2_score
'''

from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train) # Cal mean and std deviation, cal w.r.t each data points and assign
X_test=sc.transform(X_test) # when we add 'transform()', it will take mean and std deviation from X_train, which was calculated in above step
```

```python
X_train
```

#### Import Ada Boost Regressor

```python
from sklearn.ensemble import AdaBoostRegressor
from sklearn.datasets import make_regression
X, y = make_regression(n_features=5, n_informative=2,
                       random_state=0, shuffle=False)
regr = AdaBoostRegressor(random_state=0, n_estimators=100)
regr.fit(X, y)
```

```python
y_pred=regr.predict(X_test)
```

### Validation

```python
from sklearn.metrics import r2_score
r_score=r2_score(y_test,y_pred)
r_score
```
